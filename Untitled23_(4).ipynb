```python
# Data and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ML models
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier

# ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping

# SMOTE for imbalance
from imblearn.over_sampling import SMOTE

import warnings
warnings.filterwarnings('ignore')
```

```python
# --- Preprocessing: Convert 'Sleep Duration' to Numerical ---

# First, let's inspect the unique values in the column
print("Original unique values in 'Sleep Duration':")
print(df['Sleep Duration'].unique())

# Create a mapping dictionary to convert categories to numerical values
sleep_mapping = {
    'Less than 5 hours': 4.5,  # A value representing less than 5
    '5-6 hours': 5.5,          # The average of 5 and 6
    '7-8 hours': 7.5,          # The average of 7 and 8
    'More than 8 hours': 9.0   # A value representing more than 8
}

# Apply the mapping to the 'Sleep Duration' column
df['Sleep Duration'] = df['Sleep Duration'].map(sleep_mapping)

print("\n--- After Conversion ---")
print("Data type of 'Sleep Duration' column:", df['Sleep Duration'].dtype)
print("\nFirst 5 rows of the converted column:")
print(df[['Sleep Duration']].head())
```

```python
print("‚úÖ Null Values:\n", df.isnull().sum())
print("\n‚úÖ Duplicate Rows:", df.duplicated().sum())

# Drop duplicates if any
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(12, 6))
df.hist(bins=20, figsize=(15,10), color='skyblue', edgecolor='black')
plt.suptitle("Univariate Analysis (Histograms)", fontsize=14)
plt.show()

# For categorical columns
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col], palette='coolwarm')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("‚úÖ Shape after removing outliers:", df.shape)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[numeric_cols])
plt.title("Boxplot After Outlier Removal (IQR)")
plt.show()
```

```python
df['Depression'].value_counts()
```

```python
X = df.drop('Depression', axis=1)
y = df['Depression']
```

```python
X = pd.get_dummies(X, drop_first=True)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
sns.countplot(x=y)
plt.title("Class Distribution Before SMOTE")
plt.show()
```

```python
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_scaled, y)

sns.countplot(x=y_res)
plt.title("Class Distribution After SMOTE")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(),
    'Naive Bayes': GaussianNB(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

base_acc = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    base_acc[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(f"{name} Classification Report:\n{classification_report(y_test, y_pred)}\n")
```

```python
# üîß Reduced Hyperparameter Grid (faster GridSearch)
param_grids = {
    "Logistic Regression": {
        'C': [0.1, 1],
        'solver': ['liblinear']  # works well for small to medium datasets
    },
    "Decision Tree": {
        'criterion': ['gini'],
        'max_depth': [5, 10, None],
        'min_samples_split': [2, 5]
    },
    "Random Forest": {
        'n_estimators': [100],
        'max_depth': [10, None],
        'min_samples_split': [2, 5]
    },
    "SVM": {
        'C': [1, 10],
        'kernel': ['linear', 'rbf'],  # 'poly' often slower, less needed
        'gamma': ['scale']
    },
    "KNN": {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform']
    },
    "XGBoost": {
        'n_estimators': [100],
        'learning_rate': [0.1],
        'max_depth': [3, 5],
        'colsample_bytree': [1.0]
    }
}

# üîç Grid Search Loop (same)
tuned_results = []
for name, params in param_grids.items():
    print(f"\nüîç Tuning {name}...")
    grid = GridSearchCV(models[name], params, cv=3, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    tuned_results.append([name, accuracy_score(y_test, y_pred)])

# üìä Print tuned accuracies
print("\nüìä Hyperparameter Tuning Accuracies:")
for name, acc in tuned_results:
    print(f"{name}: {acc:.4f}")
```

```python
ann = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = ann.fit(X_train, y_train, validation_split=0.2, epochs=50, batch_size=32, callbacks=[early_stop], verbose=1)
```

```python
ann_acc = ann.evaluate(X_test, y_test, verbose=0)[1]
print("ANN Accuracy:", ann_acc)
```

```python
final_results = {**base_acc, **dict(tuned_results), 'ANN': ann_acc}
df_results = pd.DataFrame(list(final_results.items()), columns=['Model', 'Accuracy']).sort_values(by='Accuracy', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='Accuracy', y='Model', data=df_results, palette='viridis')
plt.title("Final Model Accuracy Comparison")
plt.show()

df_results
```
